```python
lst = []
lst1 = []
n = int(input("Enter number of elements : "))
for i in range(0, n):
	ele = int(input())
	lst.append(ele) 
print(lst)

for j in range(0, n):
	ele = int(input())
	lst1.append(ele) 
print(lst1)
```

```python
sum = 0
for k in range(0,n):
    x = lst[k]*lst1[k]
    sum += x
print(sum)
```
